```python
hotel_bookings['agent'].unique()
```

```python
hotel_bookings['agent'].fillna(-1, inplace = True)
hotel_bookings[hotel_bookings['num_children'].isna()] #Check Where num_childrean is NaN
hotel_bookings[hotel_bookings['country'].isna()][['hotel','is_canceled','country']]
hotel_bookings['country'].fillna('Unknown', inplace = True)
hotel_bookings.dropna(subset=['num_children'], inplace=True)
```

```python
hotel_bookings.drop(columns=['company'], inplace = True)
```

```python
hotel_bookings.dtypes
```

```python
hotel_bookings = hotel_bookings.astype({'is_canceled': 'boolean', 'is_repeated_guest': 'boolean', 'num_children': 'int', 'agent': 'int'})
hotel_bookings.dtypes
```

```python
hotel_bookings['lead_time'].unique()#unique vlaues
hotel_bookings['lead_time'].describe() #min max 
```

```python
# Example bins and labels (must match!)
bins = [0, 100, 200, 300, 400, 500, 600, 700, 800]
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800']

# Apply cut
hotel_bookings['lead_time_binned'] = pd.cut(hotel_bookings['lead_time'], bins=bins, labels=labels)

# Display the relevant columns
hotel_bookings[['lead_time', 'lead_time_binned']]
```

```python
hotel_bookings.head()
```

```python
column_to_move = hotel_bookings.pop('lead_time_binned')
hotel_bookings.insert(3, 'lead_time_binned', column_to_move)
hotel_bookings.head()
```

```python
hotel_bookings['hotel'].unique()
```

```python
hotel_bookings['hotel'] = hotel_bookings['hotel'].replace(r"[\*\n\^]", '', regex=True)
hotel_bookings['hotel'].unique()
```

```python
hotel_bookings.loc[hotel_bookings.duplicated(keep=False)] #remove duplicates
hotel_bookings.drop_duplicates(keep="first", inplace=True)
```

```python
hotel_bookings.loc[hotel_bookings.duplicated(keep=False)] #it will return nothing so no duplicates
```

```python
hotel_bookings.to_csv('hotel_bookings_cleaned.csv', 
          index=False,       
          encoding='utf-8-sig')
print("Successfully saved cleaned data to 'hotel_bookings_cleaned.csv'")
```
